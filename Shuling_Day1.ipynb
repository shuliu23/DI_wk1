```python
import requests
from bs4 import BeautifulSoup
from collections import namedtuple

url = "http://www.newyorksocialdiary.com/party-pictures"

# define a named tuple to store all the information
EvtBasic = namedtuple('EvtBasic', 'title, url, date')
# write a function to extract title, url, date
def td_infor(td):
    ls = td.select('a')[0]
    ldt = td.select('span.views-field-created span.field-content')[0]
    return EvtBasic(title=ls.text, url=ls['href'], date=ldt.text)

pg_EvtBasic = [None] * 25

for i in xrange(25): 
    response = requests.get(url, params={"page": i}) # change pages here
    # print response.url
    # print response.text

    soup = BeautifulSoup(response.text)
    # print soup.prettify()
    evt_divs = soup.select('div.view-content div.views-row')
    ##### Start from evt_divs
    pg_EvtBasic[i] = [td_infor(td) for td in evt_divs]

```
